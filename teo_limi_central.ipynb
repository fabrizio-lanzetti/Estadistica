# Teorema del limite central.
##### Independientemente de la forma de la distribucion de la poblacion (Ver figuras de distribucion de los datos normales y sesgados). La forma de la distribucion de las muestras de la media se aproxima a una curva normal, si el tamaño de la muestra es suficientemente grande (Ver figuras de distribucion de medias de las muestras).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore', message='FigureCanvasAgg is non-interactive, and thus cannot be shown')
```

Generamos una muestra de datos de tamaño "n" con una distribucion "sesgada" <br>
Generamos una muestra de datos de tamaño "n" con una distribucion "normal"<br>

```python
def generar_datos(sesgo,tamaño):
    aleatorios = stats.skewnorm.rvs(sesgo,size=tamaño,random_state=1)
    return aleatorios

datos_sesgados = generar_datos(10,100000)
datos_normales = generar_datos(0,100000)

fig = plt.figure()
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].hist(datos_sesgados,bins=1000,alpha=0.5,color="blue")
axs[0].set_title("Datos sesgados")
axs[1].hist(datos_normales,bins=1000,alpha=0.5,color="orange")
axs[1].set_title("Datos normales")
fig.show()
```

```python
def obtener_muestras(datos,numero_de_componentes,numero_muestras):
    for i in range(numero_muestras):
        promedios = np.array([])
        for _ in range(numero_muestras):
            obt = np.random.choice(datos,numero_de_componentes).mean()
            promedios = np.append(promedios,obt)
    return promedios

muestras_sesgadas = obtener_muestras(datos_sesgados,100,1000)
muestras_normales = obtener_muestras(datos_normales,100,1000)

fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].hist(muestras_sesgadas,bins=1000,alpha=0.5,color="blue")
axs[0].set_title("Medias de las muestras sesgadas")
axs[1].hist(muestras_normales,bins=1000,alpha=0.5,color="orange")
axs[1].set_title("Medias de las muestras normales")
fig.show()
```

Vemos como cobra sentido el teorema del limite central, no importa cual sea la forma de la distribución de la poblacion, la distribucion de los promedios obtenidos va a adaptarse a una curva normal (SIEMPRE QUE TENGAMOS SUFICIENTES PROMEDIOS OSEA TENER SUFICIENTES MUESTRAS).
